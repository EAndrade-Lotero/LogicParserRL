```python
def crear_instancias(par,modo,etiqueta):
    with open(file_name, modo, encoding = 'utf-8') as f:
        for noun in nouns[:TAMANO]:
            for verb1 in intransitive_verbs[:TAMANO]:
                if 'adj' in par:
                    for adj in adjs[:TAMANO]:
                        if 'verb2' in par:
                            for verb2 in intransitive_verbs[:TAMANO]:
                                if verb1 == verb2:
                                    continue
                                oraciones = ' '.join(par)
                                oraciones = re.sub('noun',noun,oraciones)
                                oraciones = re.sub('adj',adj,oraciones)
                                oraciones = re.sub('verb1',verb1,oraciones)
                                oraciones = re.sub('verb2',verb2,oraciones)
                                datapoint = oraciones + etiqueta + '\n'
                                f.write(''.join(datapoint))
                        else:
                            oraciones = ' '.join(par)
                            oraciones = re.sub('noun',noun,oraciones)
                            oraciones = re.sub('adj',adj,oraciones)
                            oraciones = re.sub('verb1',verb1,oraciones)
                            datapoint = oraciones + etiqueta + '\n'
                            f.write(''.join(datapoint))
                else:
                    if 'verb2' in par:
                        for verb2 in intransitive_verbs[:TAMANO]:
                                if verb1 == verb2:
                                    continue
                                oraciones = ' '.join(par)
                                oraciones = re.sub('noun',noun,oraciones)
                                oraciones = re.sub('verb1',verb1,oraciones)
                                oraciones = re.sub('verb2',verb2,oraciones)
                                datapoint = oraciones + etiqueta + '\n'
                                f.write(''.join(datapoint))
                    else:
                        oraciones = ' '.join(par)
                        oraciones = re.sub('noun',noun,oraciones)
                        oraciones = re.sub('verb1',verb1,oraciones)
                        datapoint = oraciones + etiqueta + '\n'
                        f.write(''.join(datapoint))


```

```python
inicial = True
for par in parejas:
    if inicial:
        modo = 'w'
        inicial = False
    else:
        modo = 'a'
    crear_instancias(par,modo,', 1')
```
